# Static Scatter Plot with Red Wine Quality, Alchohol, and Volatile Acidity

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(10, 15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_red['alcohol'], df_red['volatile acidity'], df_red['quality'], c=df_red['quality'], cmap='viridis')
ax.view_init(6, 35)
ax.set_xlabel('Alcohol (%)')
ax.set_ylabel('Volatile Acidity (g/L)')
ax.set_zlabel('Quality')
plt.title('Telmen Enkhtuvshin + 3D Scatter of Red Wine Quality')

# Save the plot as an image file
plt.savefig('Telmen_Enkhtuvshin+3DScatterRed_plot.png', dpi=300, bbox_inches='tight')

plt.show()
```

#### Based on this graph, we can see that Volatile Acidity and Red Wine Quality are correlated with each other. Meaning that Volatile Acidity indicates spoilage of the wine and low Volatile Acidity correlates to fresh and high quality wine. In addtition, even though alchohol percentage does not have a strong relationship with quality, we can see that higher alchohol percentage leans towards high quality.

# Interactive Scatter Plot with White Wine Quality, Alchohol, and Volatile Acidity

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(
    x=df_white['alcohol'],
    y=df_white['volatile acidity'],
    z=df_white['quality'],
    mode='markers',
    marker=dict(
        size=5,
        color=df_white['quality'],  # Optional: color by quality
        colorscale='Viridis',
        opacity=0.8
    )
)])
fig.update_layout(
    scene=dict(
        xaxis_title='Alcohol',
        yaxis_title='Volatile Acidity',
        zaxis_title='Quality'
    ),
    title='Telmen Enkhtuvshin + 3D Scatter Plot of White Wine Quality'
)
fig.show()
# Save the plot as an image file
fig.write_image('Telmen_Enkhtuvshin+3DScatter_plot.png', scale=2)  # PNG
```

### This interactive scatter plot indicates the White Wine quality with extended interactivity to view from multiple angles. Just like the red wine quality, we can see that the three variable's relationship persists too. The lower the Volatile Acidity and higher the alcohol percentage, the higher the wine quality will be.

```python
import numpy as np

# Round values to reduce uniqueness (optional for plotting surface)
# df_red["volatile acidity"] = df_red["volatile acidity"].round(2)
# df_red["sulphates"] = df_red["sulphates"].round(2)

# Create a pivot table to form a grid for surface plot
pivot = df_red.pivot_table(values='total sulfur dioxide',
                       index='sulphates',
                       columns='free sulfur dioxide',
                       aggfunc='mean')  # Use mean quality for duplicate pairs

# Get X, Y, Z values for the surface
x = pivot.columns.values  # volatile acidity
y = pivot.index.values    # sulphates
z = pivot.values          # quality

# Create 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])

fig.update_layout(title='Telmen Enkhtuvshin + Wine Total Sulfur Dioxide Surface Plot',
                  scene=dict(
                      xaxis_title='Free Sulfur Dioxide (mg/L)',
                      yaxis_title='Sulphates (g/L)',
                      zaxis_title='Total Sulfur Dioxide (mg/L)'
                  ))

fig.show()

# Save the plot as an image file
fig.write_image('Telmen_Enkhtuvshin+3DSurface_plot.png', scale=2)  # PNG
```

### We can see from this plot that Free Sulful Dioxide, Total Sulfur Dioxide, and Sulphates levels are correlated with each other. Specifically, we can see a direct relationship between these three attributes such that generally one attribute increases when the other two increases.

# Interactive Contour Plot for White Wine Citric Acid, Fixed Acidity, and pH

```python
df_white['citric acid'] = df_white['citric acid'].round(2)
df_white['pH'] = df_white['pH'].round(2)

# Pivot to create a grid
pivot = df_white.pivot_table(values='fixed acidity',
                       index='pH',
                       columns='citric acid',
                       aggfunc='mean')

# Extract axes
x = pivot.columns.values       # citric acid
y = pivot.index.values         # pH
z = pivot.values               # fixed acidity

# Create 3D surface with contour lines
fig = go.Figure(data=[go.Surface(
    z=z,
    x=x,
    y=y,
    colorscale='Viridis',
    contours={
        "x": {"show": True, "start": x.min(), "end": x.max(), "size": 0.1},
        "y": {"show": True, "start": y.min(), "end": y.max(), "size": 0.1},
        "z": {"show": True, "start": z.min(), "end": z.max(), "size": 0.5}
    }
)])

fig.update_layout(
    title='Telmen Enkhtuvshin + 3D Contour Plot: Fixed Acidity by Citric Acid and pH',
    scene=dict(
        xaxis_title='Citric Acid (g/L)',
        yaxis_title='pH',
        zaxis_title='Fixed Acidity (g/L)'
    )
)

fig.show()
# Save the plot as an image file
fig.write_image('Telmen_Enkhtuvshin+3DContour_plot.png', scale=2)  # PNG
```

### We can see here that there a trending relationship between these three variables. When Fixed acidity and Citric acid levels go up, the pH of the wine goes up just as expected.

# Interactive Bar Plot for comparing Red and White Wine Chlorides and Residual Sugar

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Load wine datasets
red_df = pd.read_csv("winequality-red.csv", sep=';')
white_df = pd.read_csv("winequality-white.csv", sep=';')

# Combine into one DataFrame with wine type column
red_df['wine_type'] = 'Red'
white_df['wine_type'] = 'White'
wine_df = pd.concat([red_df, white_df], ignore_index=True)

# Features and groupings
features = ['chlorides', 'residual sugar', 'density']
wine_types = ['Red', 'White']
colors = {'Red': 'crimson', 'White': 'gold'}

# Compute group means
group_means = wine_df.groupby('wine_type')[features].mean()

# Bar dimensions
dx, dy = 0.3, 0.3

traces = []
for i, wine in enumerate(wine_types):
    for j, feat in enumerate(features):
        height = group_means.loc[wine, feat]
        x0, x1 = i - dx/2, i + dx/2
        y0, y1 = j - dy/2, j + dy/2

        vertices = np.array([
            [x0, y0, 0], [x1, y0, 0], [x1, y1, 0], [x0, y1, 0],
            [x0, y0, height], [x1, y0, height], [x1, y1, height], [x0, y1, height]
        ])

        faces = [[0, 1, 2], [0, 2, 3],  # bottom
                 [4, 5, 6], [4, 6, 7],  # top
                 [0, 1, 5], [0, 5, 4],  # front
                 [1, 2, 6], [1, 6, 5],  # right
                 [2, 3, 7], [2, 7, 6],  # back
                 [3, 0, 4], [3, 4, 7]]  # left
        i_faces, j_faces, k_faces = zip(*faces)

        traces.append(go.Mesh3d(
            x=vertices[:, 0],
            y=vertices[:, 1],
            z=vertices[:, 2],
            i=i_faces, j=j_faces, k=k_faces,
            color=colors[wine],
            opacity=0.85,
            name=f"{wine} - {feat}",
            hovertext=f"{wine}<br>{feat}<br>Mean: {height:.4f}",
            hoverinfo="text"
        ))

# Build and display the figure
fig = go.Figure(data=traces)
fig.update_layout(
    title="Telmen Enkhtuvshin + 3D Bar Chart: Mean Chlorides, Residual Sugar, and Density in Red vs White Wine",
    scene=dict(
        xaxis=dict(title="Wine Type", tickvals=[0, 1], ticktext=wine_types),
        yaxis=dict(title="Feature", tickvals=list(range(len(features))), ticktext=features),
        zaxis=dict(title="Mean Value")
    ),
    width=800,
    height=700
)

fig.show()
# Save the plot as an image file
fig.write_image('Telmen_Enkhtuvshin+3DBar_plot.png', scale=2)  # PNG
```

### From this plot above, we can see that the mean density between red and white wine are similar, but there is a slight difference that red wine has a bit more chlorides than white wine. But the significant difference that we can see is in the residual sugar between red and white wine. Specifically, white wine has more mean residual sugar than red wine. Therefore, this is the differentiating factor between the wine types and might explain why white wine might taste more sweeter than red wine.
